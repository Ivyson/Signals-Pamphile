```python
import sympy as sm
# define the x(t) function
u = sm.Heaviside
t = sm.symbols('t')
x = (t+6)*(u(t+6)-u(t+3))+  \
3*(u(t+3)-u(t)) + \
(u(t)-u(t-2))*(-t+3) +\
(u(t-2)-u(t-4))
sm.plot(x, (t, -8, 5))
```

```python
#  Define the second Function
h = 2*(u(t-2) -u(t-3))+ (u(t-3) - u(t-4))
sm.plot(h.subs(), (t, 0, 5))
```

```python
# Convolve the two functions
tau = sm.symbols('tau')
y = sm.integrate(x.subs(t,tau)*h.subs(t, 4-tau),(tau, 0 , 7))
sm.plot(y, (t, -8, 10))
```
