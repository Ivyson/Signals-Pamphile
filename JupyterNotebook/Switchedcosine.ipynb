```python
import sympy as sm
t, omega = sm.symbols('t omega')
x = sm.Piecewise(((64*sm.cos(5*sm.pi*t)), (t > -0.3) & (t < 0.3)), (0, True))
sm.plot(x, (t, -2, 2))
y = sm.fourier_transform(x, t, omega)
absolute = sm.Abs(y)
display(sm.Eq(sm.Function("H")(omega),y))
result = absolute.subs(omega, 6)
result.evalf()
```
