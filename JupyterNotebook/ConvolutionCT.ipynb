```python
import sympy as sm
# Define the function
t = sm.Symbol('t')
x = sm.Piecewise(((t+6), (t > -6) & (t < -3)), (3, (t > -3) & (t < -1)), 
                 ((-t+1),(t > -1 ) & (t < 0)),
                 (1, (t > 0) & (t < 2)),
                 ((-t+3),(t > 2) & (t < 3)),
                 (0, True))
h = sm.Piecewise((2,(t>2) & (t < 3)), (1,(t>3) & (t < 4)), (0, True))
sm.plot(x, h.subs(t,-t+4), (t,-8,4))
result = sm.integrate(h.subs(t, 4-t)*x,(t , 0, 2))
result
```
