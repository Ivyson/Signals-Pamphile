```python
import sympy as sm

t = sm.symbols('t')
u = sm.Heaviside
h = 2*(u(t-2)-u(t-3))+(u(t-3)-u(t-4))
sm.plot(h, (t, 0, 5))
```

```python
x = 3*(u(t+3)-u(t+1))+(-t+1)*(u(t+1)-u(t))
sm.plot(x, (t, -6, 1))
```

```python
plot1 = sm.plot(h.subs(t, 2-t), (t, -5, 1), show=False, line_color="red")
plot2 = sm.plot(x, (t, -5, 1), show=False, line_color='blue')
plot1.extend(plot2)
# plot1.fill = True
plot1.show()
```

```python
result = sm.integrate(h.subs(t,2-t)*x,(t, -2, 0))
result
```
