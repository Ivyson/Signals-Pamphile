# Fourier Complex Co-efficient

$$C_k (k) = \frac{A+B}{2} = \frac{X_0}{2}$$
Whereby $A$ is the lowest point of the graph, and $B$ is the peak value of the graph. $X_0$ is the amplitude, provided that the graph begins as rest, that is: $A = 0$
