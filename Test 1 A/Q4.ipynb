For the third line starting at $x=0$ and ending at $x = 2$:

$$
m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{3 - 1}{0 - 2} = \frac{2}{-2} = -1
$$

$$

$$

For the fourth line(last) starting at $x=2$ and ending at $x=4$:

$$
y = 1
$$


```python
import sympy as sm

t = sm.Symbol('t')
y = sm.Piecewise(((t+6), (t > -6) & (t < -3)), (3, (t > -3) & (t < 0)), 
                 ((-t+3),(t > 0 ) & (t < 2)),
                 (1, (t > 2) & (t < 4)), (0, True))
sm.plot(y, (t, -7, 5))
```

```python
xeven = 0.5*(y + y.subs(t, -t))
display(f"The even function:{xeven}")
inter = sm.integrate(xeven, (t, 1, 3))
inter
```
