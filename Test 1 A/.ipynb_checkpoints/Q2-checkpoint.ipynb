Finding the complex co-efficient, we need to use the equation:

$C_k (k) = - \frac{2 X_0}{\pi^2 k^2}$

Whereby when $k=0$, then:

$C_0  = \frac{X_0}{2}$
And $X_0$ represents the peak, provided that the lowest point of the graph is at rest; aka, $y=0$, But if the graph starts somewhere in the middle, $A$ being the lowest point and $B$ being its peak, then the following equation can be used otherwise known as taking an average:

$C_0 = \frac{A + B}{2}$
