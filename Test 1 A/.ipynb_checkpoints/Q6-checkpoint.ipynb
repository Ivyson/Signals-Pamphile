```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-8, 8)

x = np.array([0, 0, 0, 1, 2, 3, 1, 1,
              1, 0, -1, -2, -2, 0, 0, 0])

fig, ax = plt.subplots(figsize=(8, 4))

markerline, stemlines, baseline = ax.stem(
    n, x,
    linefmt='C0-',
    markerfmt='C0o',
    basefmt='k-'
)


plt.setp(stemlines, linewidth=2)
plt.setp(markerline, markersize=7)

ax.set_title('Discrete-Time Signal $x[n]$', fontsize=14)
ax.set_xlabel('n', fontsize=12)
ax.set_ylabel('$x[n]$', fontsize=12)

ax.grid(True, linestyle='--', alpha=0.6)
ax.set_xticks(n)
ax.axhline(0, color='black', linewidth=1)

ax.set_xlim(-8.5, 7.5)
ax.set_ylim(-2.5, 3.5)

plt.tight_layout()
plt.show()
```

```python
# Perform time reversal and time shifting
# In Python: -1 resolves to the last value of the array, so when you
#  Use a negative index, you are starting from the end of the array
y = 3 * x[2 - n]
print(y[-2]+y[-1]+y[0]+ y[1]+y[2]+y[3]+y[4])
```
