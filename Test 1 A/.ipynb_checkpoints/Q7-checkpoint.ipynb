For the first Line from $x=-6$, to $x = -3$:

$$
m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{3 - 0}{-3 + 6} = \frac{3}{3} = 1
$$

$$
y = x + c
$$

$$
0 = -6 + c
$$

$$
c = 6
$$

$$
y = x + 6
$$

For the second Line from$x=-3$, till $x=0$:

$$
y = 3
$$

For the third line starting at $x=0$, till $x=2$:

$$
m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{3 - 1}{0 - 2} = \frac{2}{-2} = -1
$$

$$
y = -x + c
$$

$$
3 = c
$$

$$
y = -x + 3
$$

For the fourth line starting at $2$, till $x=4$:

$$
y = 1
$$


```python
import sympy as sm
from sympy.plotting import plot

t = sm.symbols('t', real=True)
H = sm.Heaviside

y = (
    (t + 6) * (H(t + 6) - H(t + 3))
    + 3 * (H(t + 3) - H(t))
    + (3 - t) * (H(t) - H(t - 2))
    + (H(t - 2) - H(t - 4))
)

h = (
    2 * (H(t - 2) - H(t - 3))
    + (H(t - 3) - H(t - 4))
)

p1 = plot(
    y, (t, -7, 5),
    label='y(t)',
    line_color='blue',
    show=False
)

p2 = plot(
    h, (t, -7, 5),
    label='h(t)',
    line_color='red',
    show=False
)

# Combine
p1.extend(p2)

p1.title = 'Signals $y(t)$ and $h(t)$'
p1.xlabel = 't'
p1.ylabel = 'Amplitude'

p1.show()
```

```python
tau = sm.Symbol('tau')
x = y.subs(t, tau)
hn = h.subs(t, t-tau)
result = sm.integrate(x*hn, (tau, -10, 10)) # -10 to 10 is  so that the integral can resolve faster than going to infinity
result.subs(t, 2)
# result
```
