For the first equation at \([-6, -3]\):

\[
m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{3 - 0}{-3 + 6} = \frac{3}{3} = 1
\]

\[
y = x + c
\]

\[
0 = -6 + c
\]

\[
c = 6
\]

\[
y = x + 6
\]

For the second equation at \([-3, 0]\):

\[
y = 3
\]

For the third equation at \([0, 2]\):

\[
m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{3 - 1}{0 - 2} = \frac{2}{-2} = -1
\]

\[
y = -x + c
\]

\[
3 = c
\]

\[
y = -x + 3
\]

For the fourth equation at \([2, 4]\):

\[
y = 1
\]


```python
import sympy as sm

t = sm.Symbol('t')
y = sm.Piecewise(((t+6), (t > -6) & (t < -3)), (3, (t > -3) & (t < 0)), 
                 ((-t+3),(t > 0 ) & (t < 2)),
                 (1, (t > 2) & (t < 4)), (0, True))
sm.plot(y, (t, -7, 5))
h = sm.Piecewise((2, (t > 2) & (t < 3)), (1, (t > 3) & (t < 4)), (0, True))
sm.plot(h,(t, -5, 5))
```

```python
tau = sm.Symbol('tau')
x = y.subs(t, tau)
hn = h.subs(t, t-tau)
result = sm.integrate(hn*t, (tau, -sm.oo, sm.oo))
# result.subs(t,2)
result.subs(t, 2)
```
