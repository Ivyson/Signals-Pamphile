```python
#
import numpy as np
import matplotlib.pyplot as plt 

time = np.arange(-10, 11) ## Creates points from -10 to 10 only integer values here..
print(time)

# Create corresponding x(t) outputs
x = np.array([0, 0, 0, 0, 0, -1, -2, -2, -1, 0, 1, 2, 2, 3, 2, 1, 0, 0, 0, 0, 0])
print(x)

plt.stem(time, x)  
plt.xlabel("Time")
plt.ylabel("x(t)")
plt.show()
```

```python
# Find the even component of this graph
xeven = 0.5 * (x + x[::-1]) # -> x[::-1] performs a time reversal here .. 
print(xeven)
t = np.where((time >= 0) & (time <= 4))
sum = xeven[t].sum()
print(sum)
```
