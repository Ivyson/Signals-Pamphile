```python
import sympy as sm
t , s = sm.symbols('t s')
vin = 8*sm.exp(-0.5*t)
R = 5
C = 0.5
# Provided below is an equation for a transfer function of an RC function.
H = 1 / (R * C * s + 1)
```

```python
Vin = sm.laplace_transform(vin, t, s)
Vout = H*Vin[0]
Vout
```

```python
vout = sm.inverse_laplace_transform(Vout, s, t)
vout
```

```python
vout.subs(t, 7.59)
```
